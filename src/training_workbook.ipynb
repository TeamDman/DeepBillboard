```python
import cv2
import common
import input_reader
import numpy as np
```

```python
images, \
image_size, \
start_points, \
occl_sizes = input_reader.get_input_images("../resources/digital_Dave_curve1/", extension="jpg")
```

```python
for image, start, occl in zip(images, start_points, occl_sizes):
    image = image[0]
    image = cv2.resize(image, (0,0), fx=1, fy=1)

    start = np.asarray(start)
    occl = np.asarray(occl)
    print(image.shape, start, occl, start+occl)

    cv2.circle(image, tuple(start), 5, (0,255,0), -1)
    cv2.circle(image, tuple(start + occl), 5, (255,0,0), -1)

    scale = 3
    image = cv2.resize(image, (0,0), fx=scale, fy=scale)
    common.display_image(image)
```
