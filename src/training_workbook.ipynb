```python
import cv2
import common
import input_reader
import numpy as np
```

```python
images, \
image_size, \
start_points, \
occl_sizes = input_reader.get_input_images("../resources/digital_Dave_curve1/", extension="jpg")
```

```python
scale = 3
for image, start, occl in zip(images, start_points, occl_sizes):
    image = image[0]
    image = cv2.resize(image, dsize=tuple(np.asarray(image.shape[:2])*scale))

    start = np.asarray(start) * scale
    occl = np.asarray(occl) * scale

    cv2.circle(image, tuple(start), 5, (0,255,0), -1)
    cv2.circle(image, tuple(start + occl), 5, (255,0,0), -1)
    print("start",start)
    print("occl", occl)
    common.display_image(image)
```
